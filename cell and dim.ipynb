```python
import csv
from itertools import groupby
from operator import itemgetter

import process_spreadsheet

afile='/data/local/hadju/dreq/matthew.csv'
with open(afile, 'r') as fi:
    reader = csv.DictReader(fi)
    records = list(reader)
    

def include(sheet_name):
    return 'fx' not in sheet_name and not sheet_name.lower().startswith('o')


def sheet_period(sheet_name):
    result = sheet_name
    if sheet_name[0:2] == 'em':
        result = sheet_name[2:]
    elif 'prim' in sheet_name:
        result = sheet_name[4:]
    elif 'aero' in sheet_name:
        result = 'aeromon'
    return result

get_table = itemgetter('miptable')
get_period = itemgetter('frequency')
get_method = itemgetter('cell_methods')

for ttable, entries in groupby(sorted(records, key=get_table), get_table):
    if include(ttable):
        print sheet_period(ttable)
        lentries = list(entries)
        periods = map(get_period, lentries)
        methods = map(get_method, lentries)

        process_spreadsheet.derive_time_usage_profile(sheet_period(ttable), periods, methods)
```

```python
def get_info(record):
    return get_table(record), get_period(record)
set(map(get_info, records))
```
